<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 1</h4>

Genera 1000 datos de una distribución normal con media 50 y varianza 9. Estima la media y la varianza de los datos generados. Ajusta un histograma y compara con la distribución teórica.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Ejercicio 1: Estimación de Varianza en Datos Simulados
samples_var = np.random.normal(50, np.sqrt(9), 1000)
estimated_var = np.var(samples_var)
plt.hist(samples_var, bins=30, density=True, alpha=0.6, color='g')
plt.title("Histograma de Datos Simulados (Normal)")
plt.show()
```

<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 2</h4>

Dado un conjunto de datos con una relación lineal $y = 3x + \epsilon$, donde $\epsilon \sim N(0,1)$:

- Simula 500 pares $(x, y)$ y ajusta una regresión lineal.
- Estima la varianza del error residual y compárala con la teórica.


```python
# Ejercicio 2: Regresión y Varianza de Errores
np.random.seed(42)
n_samples = 500
x = np.random.uniform(0, 10, n_samples)
epsilon = np.random.normal(0, 1, n_samples)
y = 3*x + epsilon
slope, intercept, _, _, _ = stats.linregress(x, y)
residuals = y - (slope*x + intercept)
residual_var = np.var(residuals)
print("La varianza de los errores es:",residual_var)
```

<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 4</h4>

Genera $X \sim N(5,4)$ y $Y \sim N(7,9)$:

- Calcula la varianza teórica de $Z = X + Y$.
- Simula 10000 valores de $Z$ y estima su varianza empírica.
- Compara con el valor teórico.


```python
samples_x = np.random.normal(5, np.sqrt(4), 10000)
samples_y = np.random.normal(7, np.sqrt(9), 10000)
samples_z = samples_x + samples_y
var_x, var_y = 4, 9
var_z_theoretical, var_z_empirical = var_x + var_y, np.var(samples_z)
print("Comparación de varianzas:", var_z_theoretical, "y" ,var_z_empirical)
```

<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 5</h4>

Descarga datos históricos de temperatura y:

- Calcula la media y la varianza de la temperatura diaria.
- Representa la evolución de la varianza con una ventana móvil de 30 días.
- Analiza si la varianza es estacionaria o cambia con el tiempo.

```python
np.random.seed(42)
temperaturas = np.random.normal(20, 5, 365)
rolling_var = [np.var(temperaturas[i:i+30]) for i in range(len(temperaturas) - 30)]
plt.plot(rolling_var)
plt.title("Evolución de la Varianza en Ventana Móvil de 30 Días")
plt.show()
```
