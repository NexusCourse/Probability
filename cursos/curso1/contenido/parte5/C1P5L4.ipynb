<h3>Distribución Poisson</h3> 

Una fábrica recibe pedidos a una tasa de $\lambda = 5$ pedidos por hora ($X \sim \text{Poisson}(5)$). Determina:

- La media y la varianza usando la FGM.
- La probabilidad de recibir exactamente 7 pedidos en una hora.
- Simula 1000 muestras en Python y compara los valores empíricos con los teóricos.

```python
# Ejercicio 1: Distribución Poisson y la FGM
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from scipy.special import factorial 

lambda_poisson = 5
mean_poisson = lambda_poisson
var_poisson = lambda_poisson
prob_7_pedidos = (lambda_poisson**7 * np.exp(-lambda_poisson)) / factorial(7)
samples_poisson = np.random.poisson(lambda_poisson, 1000)
empirical_mean_poisson = np.mean(samples_poisson)
empirical_var_poisson = np.var(samples_poisson)

plt.hist(samples_poisson, bins=15, alpha=0.6, color='g', density=True)
plt.title("Histograma de Pedidos por Hora (Poisson)")
plt.show()
```

<h3>Distribución Binomial</h3>

Un jugador tiene probabilidad $p = 0.4$ de ganar un juego. Si juega $n = 10$ veces:

- Usa la FGP para obtener la media y la varianza.
- Calcula la probabilidad de ganar exactamente 6 veces.
- Simula 5000 repeticiones y compara los resultados con la teoría.

```python
# Ejercicio 2: Distribución Binomial y la FGP
n, p = 10, 0.4
mean_binomial, var_binomial = n * p, n * p * (1 - p)
prob_6_wins = stats.binom.pmf(6, n, p)
samples_binomial = np.random.binomial(n, p, 5000)
empirical_mean_binomial, empirical_var_binomial = np.mean(samples_binomial), np.var(samples_binomial)

plt.hist(samples_binomial, bins=15, alpha=0.6, color='b', density=True)
plt.title("Histograma de Juegos Ganados (Binomial)")
plt.show()
```

```python
# Ejercicio 3: Distribución Exponencial y la FGM
lambda_exp = 1 / 2
var_exp = 1 / lambda_exp**2
prob_more_5 = 1 - stats.expon.cdf(5, scale=1/lambda_exp)
samples_exp = np.random.exponential(1/lambda_exp, 10000)
empirical_var_exp = np.var(samples_exp)

plt.hist(samples_exp, bins=15, alpha=0.6, color='r', density=True)
plt.title("Histograma de Tiempos entre Llegadas (Exponencial)")
plt.show()
```
