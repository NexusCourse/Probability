<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 1</h4>

<h3>Movimiento Browniano con Deriva</h3> 

Simula una trayectoria de movimiento browniano con deriva $\mu = 0.5, \sigma = 1$ en [0,1] con 500 pasos y grafica la trayectoria.

```python
import numpy as np
import matplotlib.pyplot as plt

T = 1.0
N = 500
dt = T / N
t = np.linspace(0, T, N + 1)
np.random.seed(42)


mu = 0.5
sigma = 1.0
dW = np.sqrt(dt) * np.random.randn(N)
W = np.concatenate(([0], np.cumsum(dW)))
X = sigma * W + mu * t

plt.figure(figsize=(10, 5))
plt.plot(t, X, label=r'$X(t) = \sigma B(t) + \mu t$')
plt.title('Ejercicio 1: Movimiento Browniano con Deriva')
plt.xlabel('Tiempo')
plt.ylabel('X(t)')
plt.grid()
plt.legend()
plt.show() 
```

<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 2</h4>

<h3>Movimiento Browniano Geométrico</h3>

Simula 1000 trayectorias de movimiento browniano estándar en el intervalo [0,1] con 500 pasos y grafica la distribución del valor final X(1). Verifica que sigue una distribución normal N(0,1).

```python
T = 1.0
N = 500
dt = T / N
t = np.linspace(0, T, N + 1)
np.random.seed(42)

mu = 0.05
sigma = 0.2
X0 = 100
dW = np.sqrt(dt) * np.random.randn(N)
W = np.concatenate(([0], np.cumsum(dW)))
Y = mu * t + sigma * W
X_geo = X0 * np.exp(Y)

plt.figure(figsize=(10, 5))
plt.plot(t, X_geo, label='Movimiento Browniano Geométrico')
plt.title('Ejercicio 2: Movimiento Browniano Geométrico')
plt.xlabel('Tiempo')
plt.ylabel('X(t)')
plt.grid()
plt.legend()
plt.show() 
```

<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 3</h4>

<h3>Movimiento Browniano con deriva y geométrico</h3>

Simula 1000 trayectorias de movimiento browniano geométrico con $\mu = 0.1, \sigma = 0.3, X(0) = 50$ y calcula el valor esperado empírico de X(1). Compara con la fórmula teórica.

```python
T = 1.0
N = 500
dt = T / N
t = np.linspace(0, T, N + 1)
np.random.seed(42)

mu = 0.1
sigma = 0.3
X0 = 50
M = 1000
X_T = np.zeros(M)

for i in range(M):
    dW = np.sqrt(dt) * np.random.randn(N)
    W = np.cumsum(dW)
    Y = mu * T + sigma * W[-1]
    X_T[i] = X0 * np.exp(Y)

# Valor esperado empírico
mean_empirical = np.mean(X_T)

# Valor esperado teórico
mean_theoretical = X0 * np.exp(mu * T)

print(f"Valor esperado empírico de X(1): {mean_empirical:.4f}")
print(f"Valor esperado teórico de X(1): {mean_theoretical:.4f}")
```
