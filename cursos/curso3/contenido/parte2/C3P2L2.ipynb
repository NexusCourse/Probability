<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 1</h4>


```python
import numpy as np

# Parámetros
n_simulations = 1000
target = 5  # Umbral de paro (+5 o -5)

# Función para simular una caminata aleatoria y calcular el tiempo de paro
def simulate_random_walk(target):
    position = 0
    steps = 0
    while abs(position) < target:
        position += np.random.choice([-1, 1])  # Movimiento aleatorio
        steps += 1
    return steps, position

# Simulación de múltiples trayectorias
stopping_times = []
final_positions = []

for _ in range(n_simulations):
    steps, final_position = simulate_random_walk(target)
    stopping_times.append(steps)
    final_positions.append(final_position)

# Cálculo de la probabilidad de alcanzar +5 antes que -5
prob_positive = np.mean(np.array(final_positions) == target)
print(f"Probabilidad de alcanzar +5 antes que -5: {prob_positive:.4f}")

# Cálculo del tiempo promedio de paro
average_stopping_time = np.mean(stopping_times)
print(f"Tiempo promedio de paro E[T]: {average_stopping_time:.2f}")

# Visualización de los tiempos de paro
import matplotlib.pyplot as plt

plt.hist(stopping_times, bins=30, edgecolor='black')
plt.title("Distribución de los Tiempos de Paro")
plt.xlabel("Tiempo de Paro (T)")
plt.ylabel("Frecuencia")
plt.show()
```

<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 2</h4>


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
n_simulations = 1000
initial_price = 100
upper_threshold = 120
lower_threshold = 80
delta = 0.03  # Variación diaria (3%)

# Función para simular una trayectoria del precio y calcular el tiempo de paro
def simulate_stock_price(initial_price, upper_threshold, lower_threshold):
    price = initial_price
    steps = 0
    while lower_threshold < price < upper_threshold:
        # Generar movimiento +3% o -3% con igual probabilidad
        movement = np.random.choice([1 + delta, 1 - delta])
        price *= movement
        steps += 1
    return steps, price

# Simulación de múltiples trayectorias
stopping_times = []
final_prices = []

for _ in range(n_simulations):
    steps, final_price = simulate_stock_price(initial_price, upper_threshold, lower_threshold)
    stopping_times.append(steps)
    final_prices.append(final_price)

# Cálculo de la probabilidad de alcanzar 120 antes que 80
prob_upper = np.mean(np.array(final_prices) >= upper_threshold)
print(f"Probabilidad de alcanzar 120 antes que 80: {prob_upper:.4f}")

# Cálculo del tiempo promedio de paro
average_stopping_time = np.mean(stopping_times)
print(f"Tiempo promedio de paro E[T]: {average_stopping_time:.2f}")

# Visualización de los tiempos de paro
plt.hist(stopping_times, bins=30, edgecolor='black', color = 'darkolivegreen')
plt.title("Distribución de los Tiempos de Paro")
plt.xlabel("Tiempo de Paro (T)")
plt.ylabel("Frecuencia")
plt.show()
```
