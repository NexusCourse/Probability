<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 1</h4>

Para definir la primera Cadena de Markov, se describe el siguiente escenario:
Un robot de servicio se encuentra en un hotel como ayudante general, para lo cual se
precisa conocer la ubicación de este cada minuto, lo cual está definido en este primer
modelo únicamente con las siguientes localidades:
 * a = Lobby
 * b = Cocina



Definiremos la matriz de probabilidad para $a$ y $b$, en su forma matricial.

```python
import numpy as np 
M = np.matrix([[0.8,0.4],[0.2,0.6]])
print(M)
```

Ahora buscamos conocer la probabilidad de que se encuentre en otra ubicación en el futuro, durante los siguientes $n$ ensayos. Al saber con certidumbre que el agente se encuentra en al ubicación  $b$. 

```python
P = np.matrix([[0],[1]])
```

Definimos la distribucion de Probabilidad como sigue:

```python
def dist_prob(T,x_0,n):
    x_n = np.matmul(np.linalg.matrix_power(M,n),x_0)
    print('P(t+'+str(n)+')=\n'+str(x_n))
```

Llamaremos a los ensayos que se requieran, en este caso hasta 6.

```python
P_7 = dist_prob(M,P,6)
```

En $P_{7}$ conocemos unicamente este valor, pero si queremos conocer los valores de $P_{1}$ a $P_{7}$, debemos realizar el siguiente ciclo for:

```python
for i in range(7):
    dist_prob(M,P,i)
    print('')
```

En el ensayo anterior, es posible reconocer que se llega  una convergencia, pero ¿Cómo saber cuando llegar a esta convergencia, sin la necesidad de calcular lo anterior?

Procedemos a Calcular eigenvectores.

```python
e,s = np.linalg.eig(M)
print(e)
print(s)
```

Identificando la columna correspondiente al eigenvalor 1, lo normalizamos.

```python
e_v = s[:,0] / sum(s[:,0])
print(e_v)
```

Note que el resultado es el mismo que en $P_{t+6}$, es decir, conocemos la convergencia sin necesidad de calcular todos los valores anteriores.

<h4 style="color: green; font-weight: bold; text-align: left;">Ejercicio 2</h4>

Ahora el robot de servicio se encuentra en este modelo dada por las siguientes
localidades:
* a = Lobby
* b = Cocina
* c = Estacionamiento



Se consideran las locaciones $a$, $b$ y $c$. La matriz de transición es:

```python
M = np.matrix([[0.6,0.1,0.2], [0.3,0.7,0.3], [0.1,0.2,0.5]])
print(M)
```

Definimos el estado inicial en $a$.

```python
P = np.matrix([[1], [0], [0]])
```

Definicion de la Distribucion de Probabilidad.

```python
def dist_prob(T,x_0,n):
    x_n = np.matmul(np.linalg.matrix_power(M,n),x_0)
    print('P(t+'+str(n)+')=\n'+str(x_n))
```

```python
P_7 = dist_prob(M,P,6)
```

```python
for i in range(7):
    dist_prob(M,P,i)
    print('')
    
```

Ahora, verificamos la convergencia.

```python
e,s = np.linalg.eig(M)
e_v = s[:,0] / sum(s[:,0])
print(e_v)
```
